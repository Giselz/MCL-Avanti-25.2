### Tema: Conceitos básicos de machine learning.

#####  1. Explique, com suas palavras, o que é machine learning?


É uma área dentro de Inteligência Artificial em que os computadores aprendem com dados, analisando grandes conjuntos de dados para assim reconhecer padrões e fazer previsões ou decisões com base em dados. Com isso, a máquina aprende a partir de exemplos.

##### 2. Explique o conceito de conjunto de treinamento, conjunto de validação e conjunto de teste em machine learning.


Esses termos se referem a prática de dividir o conjunto de dados em três subconjuntos (treinamento, validação e teste) para permitir a medição da performance de algoritmos de previsão e classificação.
O conjunto de treinamento é usado para ensinar o modelo, o conjunto de validação para ajustar hiperparâmetros e avaliar o desempenho durante o treinamento, e o conjunto de teste para avaliar o desempenho final do modelo em dados não vistos. 

##### 3. Explique como você lidaria com dados ausentes em um conjunto de dados de treinamento.


Primeiramente tentaria entender como está a distribuição de valores ausentes, sendo menor que 5% optaria pela remoção dos mesmos, pois não iriam ter um impacto forte na análise e ao mesmto tempo é uma técnica mais simples de aplicar. Caso a presença de dados ausente seja maior usaria o método de imputação usando a média ou mediana. Ao assumir que os dados são aleatoriamente ausentes, a média ou a mediana podem estar próximas dos valores originais, mas isso para valores numérios, no caso de categorias usaria a inputação da categoria mais frequente, logo a moda.
